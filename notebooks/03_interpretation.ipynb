### 1. Interpretasi Confusion Matrix & ROC Curve (XGBoost)

Berdasarkan visualisasi *Confusion Matrix*, model XGBoost menunjukkan performa klasifikasi yang sangat baik:
* **Keberhasilan Sempurna:** Model memiliki akurasi 100% pada kelas `Tata_Tertib_Kedisiplinan` (memprediksi 8 dari 8 data dengan benar).
* **Kesalahan Prediksi (Misclassification):** Terdapat sedikit irisan pada kelas `Kurikulum_Akademik`, di mana 2 pertanyaan diprediksi secara keliru sebagai Tata Tertib. Hal ini wajar dalam pemrosesan teks jika terdapat penggunaan kosa kata pengguna yang tumpang tindih antar konteks.

Pada grafik *Multiclass ROC Curve*, model menunjukkan kemampuan pemisahan kelas (*separability*) yang luar biasa dengan nilai *Area Under Curve* (AUC) yang sangat tinggi:
* `Kurikulum_Akademik` (AUC = 0.97)
* `Tata_Tertib_Kedisiplinan` (AUC = 0.95)
* `Pedoman_Tugas_Akhir` (AUC = 0.84)

### 2. Interpretasi Feature Importance

Analisis *Feature Importance* memvalidasi bahwa model XGBoost beroperasi secara rasional (tidak seperti *black box* acak). Model menggunakan fitur (kosa kata) yang sangat logis sebagai dasar pengambilan keputusan:
* Kelas **Pedoman Tugas Akhir** sangat dipengaruhi oleh keberadaan kata kunci seperti **'daftar'**, **'skripsi'**, dan **'lembar'**.
* Kelas **Kurikulum Akademik** merespons kuat terhadap kata **'
* matkul'**, **'sks'**, **'belajar'**, dan **'numerik'**.
* Kelas **Tata Tertib Kedisiplinan** banyak terbantu oleh kata bernuansa operasional kampus seperti **'kampus'** dan **'saat'**.
